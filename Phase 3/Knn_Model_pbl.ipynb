```python
#Importing libraries
import pickle
import pandas as pd

```

```python
# load model
with open("KNN_model.pkl", "rb") as f:
    model = pickle.load(f)

print("Model loaded successfully")
```

```python
#Feature names
Features=['Time', 'V3', 'V4', 'V6', 'V7', 'V8', 'V10', 'V11', 'V12', 'V14', 'V16', 'V17', 'V18', 'V19', 'V22']
print(Features)
```

```python
# take user input
values = input("Enter above mentioned features: ")
```

```python
# convert to float list
features = list(map(float, values.split()))
```

```python
# convert to DataFrame
input_df = pd.DataFrame([features], columns=Features)
```

```python
# predict
prediction = model.predict(input_df)
if (prediction==0):
    print("0:Legitimate Transaction")
else:
    print("1:Fraudulent Transaction")
```
